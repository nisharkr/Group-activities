## 1. Complete the following code to find the area of an equilateral triangle. Output should be as displayed

```python
import math
side = float(input("Enter the side of the equilateral triangle: "))
area = ((math.sqrt(3))/4)*pow(side,2)
print("Area of the equilateral triangle with side " ,side ," is ","%.2f"%area)
```

## 2. Write a program to count the number of each characters in a string

```python
strGetchar = input("Enter the string to count the characters ")
lstChar={}
for char in strGetchar:
    if char.upper() in lstChar:
        lstChar[char.upper()] +=1
    else:
        lstChar[char.upper()]= 1
print ("Number of characters in ",strGetchar," is ",lstChar)
```

## Write a program to find the area and perimeter of a rectangle using functions

```python
l=float(input("Enter the length of the rectangle :"))
b=float(input("Enter the breadth of the rectangle :"))
def rectArea(l,b):
    return round ((l*b),2)
def rectPerimeter(l,b):
    return round((2*(l+b)),2)

print ("Area of the rectangle is :",rectArea(l,b))

print ("Perimeter of the rectangle is :",rectPerimeter(l,b))
```

## 7. Complete the following code to convert hour into seconds. Output should be as displayed

```python
def to_seconds(t):
    t=round(t*60*60,2)
    return t
time_in_hours = float(input("Enter time one hours to convert into seconds:"))
print(time_in_hours ," Hour is equal to" ,to_seconds(time_in_hours) ," Seconds")
```
